```python
import os
import pandas as pd
```

```python
call = pd.read_csv('Call.csv', encoding='cp949')
interestdiff = pd.read_csv('InterestDiff.csv', encoding='cp949')
vkospi = pd.read_csv('VKOSPI.csv', encoding='cp949')
tradbal = pd.read_excel('TradeBalance.xlsx')
cpi = pd.read_excel('CPI.xlsx')
ppi = pd.read_csv('PPI.csv', encoding='cp949')
unemploy = pd.read_excel('Unemployment.xlsx')
manuprod = pd.read_excel('ManufactProduct.xlsx')
spread = pd.read_csv('SPREAD.csv', encoding='cp949')
erlog = pd.read_csv('Erlog.csv', encoding='cp949')
```

```python
spread.info()
```

```python
# datetime 변환
call['시점'] = pd.to_datetime(call['시점'])
interestdiff['시점'] = pd.to_datetime(interestdiff['시점'])
vkospi['시점'] = pd.to_datetime(vkospi['시점'])
tradbal['시점'] = pd.to_datetime(tradbal['시점'])
cpi['시점'] = pd.to_datetime(cpi['시점'])
ppi['시점'] = pd.to_datetime(ppi['시점'])
unemploy['시점'] = pd.to_datetime(unemploy['시점'])
manuprod['시점'] = pd.to_datetime(manuprod['시점'])
spread['시점'] = pd.to_datetime(spread['시점'])
erlog['시점'] = pd.to_datetime(erlog['시점'])
```

```python
# datetime 변환 확인
call.info()
interestdiff.info()
vkospi.info()
tradbal.info()
cpi.info()
ppi.info()
unemploy.info()
manuprod.info()
spread.info()
```

## 병합

### 1. VKOSPI 포함

```python
# 병합
a = pd.merge(call, interestdiff, how='outer')
a = pd.merge(a, vkospi, how='outer')
a = pd.merge(a, tradbal, how='outer')
a = pd.merge(a, cpi, how='outer')
a = pd.merge(a, ppi, how='outer')
a = pd.merge(a, unemploy, how='outer')
a = pd.merge(a, manuprod, how='outer')
a = pd.merge(a, spread, how='outer')
a = pd.merge(a, erlog, how='outer')
a
```

```python
# 기간 조정
# VKOSPI: 2013.08.06 ~ 2022.11.09
# 월간 변수 다수: ~2022.09
a = a[a['시점'].between('2013-08-06', '2022-09-01')]
```

```python
pd.set_option('display.min_rows', 100)
a = a.sort_values(by='시점')
a
```

```python
pd.set_option('display.min_rows', 30)
a = a.fillna(method='bfill')
```

```python
total = a
```

```python
total.isna().sum()
```
